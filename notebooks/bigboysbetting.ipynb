```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
files = []
for file in os.listdir("data/"):
    files.append(pd.read_csv("data/" + file, parse_dates=["Date"], dayfirst=True))
```

```python
df = pd.concat(files)
```

```python
big_teams_list = ["Man United", "Bayern Munich", "Man City", 
                  "Barcelona", "Real Madrid", "Juventus",
                  "Paris SG", "Celtic", "Olympiakos"]
```

```python
big_teams_df = df[(df.HomeTeam.isin(big_teams_list) & (~df.AwayTeam.isin(big_teams_list)))][["HomeTeam", "FTR", "Date", "BbMxH", "B365H"]].sort_values("Date")
```

```python
big_teams_df.head()
```

```python
len(big_teams_df)
```

```python
bankroll = 1000
history = []

for row in big_teams_df.itertuples():
    # spend 10% of money on each match
    stake = bankroll * 0.1
    bankroll -= stake
    if row.FTR == "H":
        bankroll += row.BbMxH*stake
    history.append(bankroll)
pd.Series(history).plot()
plt.show()
bankroll
```
